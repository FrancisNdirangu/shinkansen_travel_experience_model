```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5,weights='distance')
```

```python
cat_cols = final_traindf.select_dtypes('object')
cat_cols.head()
```

```python
cat_cols.Customer_Type.unique()
```

All the categorical columns other than gender are ordinal. 
They have a heirarchy of order to them meaning that we have to use ordinal/label encoding


```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()
enc_cols = enc.fit_transform(cat_cols)
enc_cols = pd.DataFrame(data=enc_cols,columns=enc.get_feature_names_out(cat_cols.columns.tolist()))
enc_cols.head()
```

```python
enc_cols.isna().any().sort_values(ascending=False)
```

Perfect I expect there to still be nan valies. I will now use the knn imputer to impute the missing values

```python
enc_cols_imputed = imputer.fit_transform(enc_cols)
enc_cols_imputed.shape
#
df_cat_imputed = pd.DataFrame(data=enc_cols_imputed,columns=enc_cols.columns.tolist())
df_cat_imputed.isna().sum().sort_values(ascending=False).head()
```

```python
df_cat_imputed.isna().any()
```

```python
num_cols  = final_traindf.select_dtypes('number').drop('ID',axis=1)
num_cols.info()
```

```python
num_cols.isna().any()
```

We will begin my minmaxscaling the numerical cols then we will use knnimputer to impute the values.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_scaled = scaler.fit_transform(num_cols)
num_scaled_df = pd.DataFrame(data=num_scaled,columns=scaler.get_feature_names_out(num_cols.columns.tolist()))
num_scaled_df.head()
```

```python
num_cols_imputed = imputer.fit_transform(num_scaled_df)
num_imputed_df = pd.DataFrame(data=num_cols_imputed,columns=imputer.get_feature_names_out(num_scaled_df.columns.tolist()))
num_imputed_df.head()
```

```python
num_imputed_df.isna().any()
```
